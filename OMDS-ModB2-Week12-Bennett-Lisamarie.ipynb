# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

**My Answer**

Side-by-Side Plot (First Image)
Advantages:

Keeps the two weeks visually separate, which makes it easier to compare trends within each week individually.

Reduces overlap confusion and avoids misinterpretation due to visual crowding.

Disadvantages:

Harder to directly compare specific days across weeks (e.g., Monday to Monday) since theyre in different plots.

Takes up more space and might not highlight overall week-to-week changes clearly.

Overlay Plot (Second Image)
Advantages:

Makes it easier to directly compare visitor counts on the same day of the week between the two weeks.

More compact and visually efficient for detecting differences between weeks.

Disadvantages:

Lines may overlap and become confusing, especially if the patterns are very similar.

Can be harder to focus on each week independently.

```python
# Set the style
sns.set_style("whitegrid")
# Data for products AE
years = list(range(2007, 2015))
product_A = [400, 420, 430, 430, 400, 290, 270, 260]
product_B = [370, 400, 410, 415, 390, 280, 260, 250]
product_C = [0, 0, 0, 100, 150, 200, 180, 220]
product_D = [0, 0, 0, 0, 140, 190, 200, 190]
product_E = [0, 0, 0, 0, 0, 100, 150, 200]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(years, product_A, color='black', linewidth=2)
plt.plot(years, product_B, color='black', linewidth=2)
plt.plot(years, product_C, color='black', linewidth=2)
plt.plot(years, product_D, color='black', linewidth=2)
plt.plot(years, product_E, color='black', linewidth=2)
# Simplified style for "eliminate clutter"
for label, x, y in zip(['A', 'B', 'C', 'D', 'E'],
                       [2007, 2007, 2010, 2011, 2013],
                       [400, 370, 100, 140, 100]):
    plt.text(x, y, label, fontsize=12)

plt.title("Average Retail Product Price per Year", fontsize=14)
plt.ylabel("$")
plt.xticks(years)
plt.yticks(range(0, 501, 100))
plt.ylim(0, 500)
plt.grid(False)
plt.box(False)
plt.tick_params(left=False, bottom=False)
plt.tight_layout()

plt.show()
```
